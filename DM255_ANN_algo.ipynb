```python
import requests
from io import StringIO
import pandas as pd
data = pd.read_csv('/content/Kaggel_news_data.csv')
data.head()
```

```python
Headline = data['Headline'].tolist()
len(set(Headline)) 
```

```python
# remove duplicates and NaN
Headline = [word for word in list(set(Headline)) if type(word) is str]
```

```python
from sentence_transformers import SentenceTransformer

# initialize Headlines transformer model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# create headline embeddings
headline_embeddings = model.encode(Headline)
headline_embeddings.shape
```
